```python
df1['PLAYER NAME ']=df1['PLAYER NAME'].str.strip()
df['PLAYER NAME ']=df['PLAYER NAME'].str.strip()
```

```python
mask = df1['PLAYER NAME'].isin(df['PLAYER NAME'])
result_df = df1[~mask]
rd=pd.DataFrame(result_df)
print(rd)
```

```python
df2 = rd.T.drop_duplicates().T
df2
```

```python
df2 = df2.loc[df1['COUNTRY'] == "INDIA"]
```

```python
df=df2.dropna()
df
```

```python
X=df.drop(['PLAYER NAME','TEAM NAME','COUNTRY','Overall','Potential'], axis=1)
Y=df[['Overall','Potential']]
X
```

```python
x_train, x_test,y_train,y_test = train_test_split(X,Y,test_size =0.15,random_state=0)
# print the data
x_train
```

```python
from catboost import CatBoostRegressor
```

```python
nan_df = df2[df2['Overall'].isna()]
```

```python
nan_df
```

```python
# Initialize CatBoost Regressor
model = CatBoostRegressor(objective='MultiRMSE')

# Train the model
model.fit(x_train,y_train)

# Predict the missing values
nan_df[['PredO','PredP']] = model.predict(nan_df.drop(['PLAYER NAME','TEAM NAME','COUNTRY','Overall','Potential'], axis=1))
```

```python
nan_df['Overall']=nan_df['PredO']
nan_df['Potential']=nan_df['PredP']
nan_df
```

```python
mask1 = df2['PLAYER NAME'].isin(nan_df['PLAYER NAME'])

# Use numpy.where with the mask to conditionally replace values in df1
df2.loc[mask1, 'Overall'] = nan_df['Overall']
df2.loc[mask1, 'Potential'] = nan_df['Potential']
```
